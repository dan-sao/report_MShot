# choose params

```python
R_min = 90
Depth_min = 1.25
```

```python
#drop non closed offers
rep_1.dropna(inplace = True)

#form sibsets with MIN-rules
rep_mins = rep_1.loc[(rep_1.R >= R_min) & (rep_1.Depth >= Depth_min)]
rep_Rmin = rep_1.loc[(rep_1.R >= R_min)]
rep_Dmin = rep_1.loc[(rep_1.Depth >= Depth_min)]

#form analytical df
adf = pd.DataFrame(columns = ['Coin', '%RB(+)', 'Depth_average'])
adf.Coin = rep_1.Coin.unique()

for i in adf.Coin:
    #quantity of deals where Depth >= Depth_min / quantity of all deals
    adf['%RB(+)'].loc[(adf.Coin == i)] = len(rep_Dmin.loc[(rep_Dmin.Coin == i)]) / len(rep_1.loc[(rep_1.Coin == i)])
    
    #average on basis of deals where Depth >= Depth_min
    adf['Depth_average'].loc[(adf.Coin == i)] = rep_Dmin[rep_Dmin.Coin == i].Depth.mean()
    
adf
```

```python
adf.Depth_average.describe()
```

```python
rep_2 = rep_1.copy()
rep_2 = rep_2.merge(adf, on = 'Coin', how = 'outer')
rep_2
```
